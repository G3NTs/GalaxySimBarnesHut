```python
import numpy
import matplotlib.animation as animation
import matplotlib.pyplot as plt
import random

from functools import partial
from IPython.display import HTML

class body_v4:
    def __init__(self, mass, radius, pos, vel, acc, force, index, deltaTime):
        self.mass = mass
        self.radius = radius

        self.pos = pos
        self.vel = vel
        self.acc = acc
        self.force = force

        self.index = index
        self.deltaTime = deltaTime

    def Update_acc(self):
        self.acc = self.force/self.mass
    def Update_vel(self):
        self.vel = self.vel + self.acc * DELTATIME
    def Update_pos(self):
        self.pos = self.pos + self.vel * DELTATIME
    def Reset_force(self):
        self.force = numpy.double([0,0,0])


def UpdateFrame(frame, List, points):
    ax.clear()
    Interaction_Force(List)
    for body_n, point in zip(List,points):
        body_n.Update_acc()
        body_n.Update_vel()
        body_n.Update_pos()
        body_n.Reset_force()
        point = ax.plot([body_n.pos[0]],[body_n.pos[1]],[body_n.pos[2]],'o',markersize=10)[0]
    return points

def Start(List):
    for i in range(10):
        x1 = random.random()-0.5
        x2 = random.random()-0.5
        x3 = random.random()-0.5
        v1 = random.random()-0.5
        v2 = random.random()-0.5
        v3 = random.random()-0.5
        List.append(body_v4(mass = 20000000000, radius = 0.1, pos = numpy.double([20*x1,20*x2,20*x3]), vel = numpy.double([2*v1,2*v2,2*v3]), acc = numpy.double([0,0,0]), force = numpy.double([0,0,0]), index = i, deltaTime=deltaTime))

def Check_Collision(magnitude, distance, body1, body2):
    if (magnitude <= (body1.radius + body2.radius)):
        new_vel1 = body1.vel - ((2*body2.mass)/(body1.mass + body2.mass)) * ((numpy.dot((body1.vel-body2.vel),(-distance)))/numpy.power(magnitude, 2))*(-distance)
        new_vel2 = body2.vel - ((2*body1.mass)/(body1.mass + body2.mass)) * ((numpy.dot((body2.vel-body1.vel),(distance)))/numpy.power(magnitude, 2))*(distance)
        body1.vel = new_vel1
        body2.vel = new_vel2
        display("collision!")

def Interaction_Force(List):
    for body1 in List:
        for body2 in List[body1.index + 1:]:
            distance = body2.pos - body1.pos
            magnitude = numpy.sqrt(distance.dot(distance))
            BodyForces = (GRAV_CONST*body1.mass*body2.mass*distance)/numpy.power(magnitude, 3)
            body1.force = body1.force + BodyForces
            body2.force = body2.force - BodyForces
            #Check_Collision(magnitude, distance, body1, body2)

DELTATIME = 0.1
GRAV_CONST = 6.674e-11

List_V4 = []
random.seed(45)
Start(List_V4)
display(List_V4)

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
points = [ax.plot([], [], [], marker='.', markersize=50)[0] for _ in List_V4]

#for _ in range(10):
#    UpdateFrame(0,List_V4,points)



anim = animation.FuncAnimation(fig, partial(UpdateFrame, List = List_V4, points = points), frames = 500, interval=100, blit=False)

plt.close(anim._fig)

HTML(anim.to_html5_video())

```
